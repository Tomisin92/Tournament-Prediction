```python
# import dependencies
import pandas as pd
import numpy as np
import numerapi
from sklearn.linear_model import LinearRegression 

import warnings 
warnings.filterwarnings("ignore")
```

```python
# operating system
import os
os.getcwd()
```

```python
training_data=pd.read_csv("latest_numerai_training_data.csv.xz")
training_data.head()
```

```python
training_data.tail()
```

```python
# Examining the data size
print(training_data.shape)
```

```python
# considering only the feature columns from the dataframe
explanatory_variables = training_data.columns[training_data.columns.str.startswith('feature')]
```

```python

# select those columns out of the training dataset
training_features = training_data[explanatory_variables]
```

```python
training_features.head()
```

```python
response_variables = training_data.columns[training_data.columns.str.startswith('target')]
```

```python
training_target = training_data[response_variables]
```

```python
training_target.head()
```

```python
# Invoking the linear regression function and find the best fit model on training data
Regression_Model=LinearRegression()
Regression_Model.fit(training_features, training_target)
```

```python
print(Regression_Model.intercept_)
```

#### The intercept of the model is 0.4802

```python
print(Regression_Model.coef_)
```

```python
# R square on training data

Regression_Model.score(training_features, training_target)
```

### Generate first prediction
Now that we have a trained model, we can use it to make predictions on the tournament data.

```python

# select the feature columns from the tournament data
live_features = tournament_data[explanatory_variables]
```

```python
# predict the target on the live features
predictions = Regression_Model.predict(live_features)          
```

```python
# predictions must have an `id` column and a `prediction` column
predictions_df = tournament_data["id"].to_frame()
predictions_df["prediction"] = predictions
predictions_df.head(5)
     
```
